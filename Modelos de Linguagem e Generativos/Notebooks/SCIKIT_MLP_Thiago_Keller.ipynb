## 1. Introdução ao Multilayer Perceptron (MLP)
O MLP é uma arquitetura de rede neural artificial feedforward que consiste em camadas de nós: entrada, ocultas e saída. Diferente de algoritmos lineares, o MLP utiliza funções de ativação não-lineares para aprender padrões complexos.

Neste projeto, utilizaremos o `MLPClassifier` para um problema de classificação multivariada, onde a rede deve identificar padrões sutis em 13 atributos diferentes para categorizar três classes distintas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Carregamento do Dataset
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target

print(f"Dataset carregado: {X.shape[0]} amostras e {len(np.unique(y))} classes.")
X.head()
```

## 2. Processamento e Normalização
Redes Neurais são sensíveis à escala dos dados. A padronização (`StandardScaler`) é essencial para que o algoritmo de otimização (Adam) converja de forma eficiente, evitando que atributos com valores altos dominem o cálculo do erro.

```python
# Divisão Treino/Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalização Obrigatória para MLP
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Dados escalonados com sucesso.")
```

## 3. Arquitetura da Rede Neural
Definimos uma rede com **duas camadas ocultas** de 10 neurônios cada. Utilizamos a função de ativação **ReLU** (Rectified Linear Unit) e o otimizador **Adam**, que é altamente eficiente para este tipo de problema.

```python
# Configuração do MLP
mlp = MLPClassifier(hidden_layer_sizes=(10, 10), 
                    max_iter=1000, 
                    activation='relu', 
                    solver='adam', 
                    random_state=42)

# Treinamento
mlp.fit(X_train_scaled, y_train)

# Visualização da Curva de Perda (Loss Curve)
plt.figure(figsize=(8, 5))
plt.plot(mlp.loss_curve_, color='purple')
plt.title("Evolução do Aprendizado (Loss Curve)")
plt.xlabel("Iterações (Épocas)")
plt.ylabel("Custo (Loss)")
plt.show()
```

## 4. Avaliação e Métricas
Abaixo, avaliamos o desempenho da rede através do Relatório de Classificação e da Matriz de Confusão Visual.

```python
predictions = mlp.predict(X_test_scaled)

print("Métricas de Performance:")
print(classification_report(y_test, predictions))

# Visualização da Matriz de Confusão
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, cmap='viridis')
plt.xlabel('Predito')
plt.ylabel('Real')
plt.title('Matriz de Confusão - MLP Neural Network')
plt.show()
```

## 5. Conclusão Final
O modelo MLP apresentou excelente poder de generalização. A curva de perda mostra uma queda acentuada nas primeiras 200 iterações, indicando uma convergência rápida e estável. A escolha de uma arquitetura multicamadas permitiu que a rede capturasse as nuances químicas dos dados com alta precisão.
